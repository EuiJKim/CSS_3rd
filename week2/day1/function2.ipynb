```python
#'add' : 두 값을 더하는 함수
#'sub' : a - b를 실행하는 함수

def calc(kind = 'add'):
    def add(a, b):
        return a + b
    def sub(a, b):
        return a - b
    
    if kind == 'add':
        return add
    elif kind == 'sub':
        return sub
    
```

```python
f = calc('add')
```

```python
f(1, 5)
```

```python
f = calc('sub')
```

```python
f(6, 2)
```
